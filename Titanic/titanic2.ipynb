```python
# Import os
import os
# Import the Pandas library
import pandas as pd
#Import NumPy as np
import numpy as np
# Import 'tree' from scikit-learn library
from sklearn.ensemble import RandomForestClassifier
# Import 'tree' from scikit-learn library
from sklearn import tree
```

```python
os.getcwd()
```

```python
# Load the train and test datasets to create two DataFrames
train_datapath = os.path.join(os.getcwd(), 'train.csv')
train = pd.read_csv(train_datapath)

test_datapath = os.path.join(os.getcwd(), 'test.csv')
test = pd.read_csv(test_datapath)
#Print the `head` of the train and test dataframes
train.head(10)
```

```python
train = train.drop(['Ticket', 'Cabin'], axis=1)
train = train.dropna()
train['Sex'] = train['Sex'].apply(lambda gender: 1 if gender == 'male' else 0)
train['Embarked'] = train['Embarked'].apply(lambda embarked: 1 if embarked == 'S' else 1 if embarked == 'C' else 0)
print(train.head(10))
```

```python
# We want the Pclass, Age, Sex, Fare,SibSp, Parch, and Embarked variables
features_forest = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
target = train["Survived"].values
```

```python
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_forest, target)
```

```python
# Building and fitting my_forest
forest = RandomForestClassifier(max_depth = 10, min_samples_split=2, n_estimators = 100, random_state = 1)
my_forest = forest.fit(features_forest, target)
```

```python
# Check performance of forest classifier
print(my_forest.feature_importances_)
print(my_forest.score(features_forest, target))
```

```python
test.head(10)
```

```python
test = test.drop(['Ticket', 'Cabin'], axis=1)
```

```python
test = test.fillna(test.mean())
test['Sex'] = test['Sex'].apply(lambda gender: 1 if gender == 'male' else 0)
test['Embarked'] = test['Embarked'].apply(lambda embarked: 1 if embarked == 'S' else 1 if embarked == 'C' else 0)
test.head(10)
```

```python
test_features_forest = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
```

```python
my_forest_prediction_1 = my_forest.predict(test_features_forest)
```

```python
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution = pd.DataFrame(my_forest_prediction_1, PassengerId, columns = ["Survived"])
print(my_solution)
```
